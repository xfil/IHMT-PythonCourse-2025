```python
entosurv['HouseLar+'] = entosurv['CasaLar +'] > 0 # transformar em boleanos a variavel CasaLar + para HouseLar+
print(entosurv.head()) # tabela com adio da coluna HouseLar+
```

```python
pivot_table = pd.pivot_table(entosurv, values='HouseLar+', index=['Localidad', 'Ubica.'], aggfunc=['sum', lambda x: (x==False).sum()]) # agg func funo de agragao com soma
pivot_table.columns = ['True_Count', 'False_Count']
pivot_table
print(pivot_table)
```

```python
larva_counts = entosurv.groupby(['Localidad', 'Ubica.'])['HouseLar+'].sum().reset_index() # variavel larva_counts; define o nr de casas + para larvas organizadas por Ubica. (local) e comuniade
print(larva_counts)
```

```python
larva_counts_pivot = larva_counts.pivot(index='Localidad', columns='Ubica.', values='HouseLar+')
print(larva_counts_pivot)
```

```python
larva_counts_pivot.plot(kind='bar', figsize=(8, 4))
plt.title('Contagem de Casas com Larvas (+) por Localidade e Ubicao')
plt.xlabel('Localidade')
plt.ylabel('Nmero de Casas com Larvas (+)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
from Bio import Entrez, SeqIO # aceder a bases de dados do GenBank (Gene Coi Ae. albopictus)
Entrez.email = "tania.reis@us.edu.cv"
```

```python
continentes=[
    "MH587188",  # sia (Japao)
    "PQ156979",  # frica (Angola)
    "MN513353",  # Europa (Portugal)
    "KX383917"   # Amrica  (Venesuela)
   ]
```

```python
sequences=[]
for paises in continentes:
    handle = Entrez.efetch(db="nucleotide", id= paises, rettype="fasta", retmode="text")
    seq = SeqIO.read(handle, "fasta")
    sequences.append(seq)
```

```python
SeqIO.write(sequences, "seqcontinent.fasta","fasta")

print (f"salvas {len(sequences)} sequencias em 'seqcontinent.fasta'")
```

# Deteco dos SNPs a partir do Counter

```python
!pip install biopython
```

```python
from collections import Counter
```

```python
snp_positions=[]
for i in range(alignment.get_alignment_length()):
    coluna = alignment[:, i]
    counts = Counter(coluna)
    if '-' in counts:
        del counts['-']
    if len(counts) > 1: # identifica diferena nos simbolos em cada coluna
      snp_positions.append(i)
```

```python
print(f"Total de SNPs encontrados: {len(snp_positions)}\n")
```

# Manipulao do alinhamento a partir do SeqRecord.

```python
from Bio.SeqRecord import SeqRecord
```

```python
half_janela = 6

for pos in snp_positions[:3]:  # mostrar s os 3 primeiros SNPs
    print(f"SNP na posio {pos+1}:\n")
    start = max(0, pos - half_janela)
    end = min(alignment.get_alignment_length(), pos + half_janela)

    janela_mutacao=[]

    for record in alignment:
        seq_trecho = record.seq[start:end]
        highlight_index = pos - start

        print(f"{record.id:20} {seq_trecho[:highlight_index]}{seq_trecho[highlight_index].upper()}{seq_trecho[highlight_index+1:]}")
    print("\n")
```

# Calculo de matrix de distncia das sequencias alinhadas a partir do mdulo Phylo.TreeConstruction

```python
from Bio.Phylo.TreeConstruction import DistanceCalculator, DistanceTreeConstructor
from Bio import Phylo
```

```python
calculator = DistanceCalculator("identity") #Cria um objeto DistanceCalculator usando o mtodo "identity".
distance_matrix = calculator.get_distance(alignment) #Calcula a matriz de distncias entre todas as sequncias do alinhamento.
```

```python
print(distance_matrix)
```

# Construo da rvore filogentica

```python
constructor = DistanceTreeConstructor()
upgma_tree = constructor.upgma(distance_matrix)
```

```python
Phylo.draw_ascii(upgma_tree)
```

```python
import matplotlib.pyplot as plt
```

```python
Phylo.draw(upgma_tree, do_show=True)
```
